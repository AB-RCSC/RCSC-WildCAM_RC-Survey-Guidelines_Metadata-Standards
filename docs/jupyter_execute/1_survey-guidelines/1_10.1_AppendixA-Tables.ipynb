**Appendix A - Table A2**

(#survey-guidelines-appendix-a-table-a-2)=

**Table A2.** Summary of appropriate [study design](#Hierarch_Survey), [camera spacing](#Camera_spacing), and [survey](#Hierarch_Survey) effort (adapted from Wearn & Glover-Kapfer [2017] with additional references included) for various [modelling approaches](#Mods_Modelling_approach).
